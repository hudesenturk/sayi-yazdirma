```python
#Sayı yazma
birler={"0":"","1":"bir","2":"iki","3":"üç","4":"dört","5":"beş","6":"altı","7":"yedi","8":"sekiz","9":"dokuz"}

onlar={"0":"","1":"on","2":"yirmi","3":"otuz","4":"kırk","5":"elli","6":"altmış","7":"yetmiş","8":"seksen","9":"doksan"}

yuzler=["yüz","bin","milyon","milyar"]

while True:
    yazi=""
    sayi=input("Üç basamaklı bir sayı giriniz: ")

    for i in range(len(sayi)):
        if i==0:
            if sayi[i] != "0":
                if sayi[i] != "1":
                    yazi+=birler[(sayi[i])]
                    yazi+=yuzler[0]
                else:
                    yazi+=yuzler[0]

            else:
                yazi+=""
        elif i==1:
            yazi+=onlar[(sayi[1])]

        else:
            yazi+=birler[(sayi[2])]

    print(yazi)

    secim=input("Devam edecek misiniz? evet/hayır ")

    if secim=="hayır":
        break
```
